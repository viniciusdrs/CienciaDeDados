```python
fifa = pd.read_csv('fifa_df.csv')
```

## Todos os Atributos dos 11 Primeiros Jogadores

```python
fifa.head(11)
```

## Apagando as Colunas Flag, Photo e Club Logo da Tabela

```python
fifa.drop(['Flag'], axis=1, inplace=True)
fifa.drop(['Club Logo'], axis=1, inplace=True)
fifa.drop(['Photo'], axis=1, inplace=True)
fifa.head()
```

## Alguns Atributos dos 11 Primeiros Jogadores do FIFA 19


```python
variaveis = ["ID", "Name", "Age", "Nationality", "Overall", "Potential", "Club"]
fifa.loc[0:10, variaveis]
```

## Todos os Jogadores Brasileiros Cujo Overall Seja Igual ou Maior que 80

```python
fifa[(fifa['Nationality'] == 'Brazil') & (fifa['Overall'] >= 80)]
```

## Gráfico mostrando o Overall dos Jogador

 Se percebe que os maiores Overall então entre 60 e 70.

```python
overall = fifa['Overall']

plt.hist(overall)

plt.show()
```

## Dispersão entre a Idade e o Valor de cada Jogador
Os jogadores de Maiores valores estão entre 25 e 35 Anos.

```python
fifa.plot(x='Value', y='Age', kind='scatter')
plt.show()
```

## Correlação entre a Idade e o Overall dos Jogadores
Se percebe que os maiores Overall estão entre os jogadores de Idade 20 e 30 Anos.

```python
sns.pairplot(fifa.loc[:,'Age':'Overall'])
plt.show()
```
